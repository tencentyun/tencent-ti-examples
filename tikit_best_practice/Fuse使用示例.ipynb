# Fuse使用示例

本示例将演示从opeml下载公开数据集，然后利用Fuse工具对该数据集进行自动特征工程处理的流程。

### 1. 下载公开数据集

#### 下载公开测试数据hill_valley，数据集来源：
Author: Lee Graham, Franz Oppacher
Source: [UCI](http://archive.ics.uci.edu/ml/datasets/hill-valley)
Please cite:

Each record represents 100 points on a two-dimensional graph. When plotted in order (from 1 through 100) as the Y coordinate, the points will create either a Hill (a “bump” in the terrain) or a Valley (a “dip” in the terrain).
See the original source for some examples of these graphs.

In the original form, there are six files. This is the non-noisy version, with training and test sets merged.

#### Attribute Information:

1-100: Labeled “V##”. Floating point values (numeric), the Y-values of the graphs.
101: Labeled “class”. Binary {0, 1} representing {valley, hill}

```python
import requests

url = 'https://www.openml.org/data/get_csv/1590101/php3isjYz'
r = requests.get(url, allow_redirects=True)
open('./valley.csv', 'wb').write(r.content)
```

### 2. 加载公开数据集

```python
import pandas as pd
file_path = "./valley.csv"
df = pd.read_csv(file_path)
df
```

### 3. 调用Fuse进行自动特征工程

```python
import fuse
fuse_af = fuse.AutoFeature()
out_df = fuse_af.fit_transform(
    df,
    index_column=None,
    label_column="Class",
)
out_df
```

### 4. 模型训练

#### 对生成特征进行建模测试
这里我们分别将原始特征和Fuse生成特征做7:3的训练/测试集切分，然后利用默认的xgboost分类器分别对二者的训练集数据进行训练，然后在测试集上进行测试，并评估模型再测试集上的AUC，用于对比Fuse生成特征的效果。

```python
import re
import numpy as np
from xgboost import XGBClassifier
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score, accuracy_score


def split_data(train_df, index, label):
    regex = re.compile(r"\[|\]|<", re.IGNORECASE)
    
    # replace some col name that model could get error
    train_df.columns = [regex.sub("_", col) if any(x in str(col) for x in set(
        ('[', ']', '<'))) else col for col in train_df.columns.values]
    label_df = train_df[label]
    selected_cols = [x for x in train_df.columns if x != index and x != label]
    input_df = train_df[selected_cols]
    X_train, X_test, y_train, y_test = train_test_split(
        input_df, label_df, test_size=0.3, random_state=11,
        stratify=label_df.to_numpy())
    return X_train, X_test, y_train, y_test


def test_auc(X_train, X_test, y_train, y_test):
    """Calculate the auc by using default XGBClassifier."""
    
    model = XGBClassifier(random_state=11)
    model.fit(X_train, y_train)
    y_prob = model.predict_proba(X_test)[:, 1]
    y_pred = np.where(y_prob >= 0.5, 1, 0)
    y_pred = model.predict(X_test)
    auc = accuracy_score(y_test, y_pred)
    return auc

def compare(raw_df, fuse_df, index, label):
    X_train, X_test, y_train, y_test = split_data(raw_df, index, label)
    raw_auc = test_auc(X_train, X_test, y_train, y_test)
    #
    X_train, X_test, y_train, y_test = split_data(fuse_df, index, label)
    fuse_auc = test_auc(X_train, X_test, y_train, y_test)
    print("########################")
    print("Test result for compare raw & fuse feature on xgboost auc.")
    print("Raw feature AUC: {}".format(raw_auc))
    print("Fuse feature AUC: {}".format(fuse_auc))
    print("########################")
```

```python
compare(df, out_df, index=None, label="Class")
```
